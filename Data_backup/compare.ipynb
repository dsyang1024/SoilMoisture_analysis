```python
import pandas as pd
# read xlsx file
FCfinder_data = pd.read_excel('./comparison_with_old_2023.xlsx', sheet_name='FC_finder')
CB_data = pd.read_excel('./comparison_with_old_2023.xlsx', sheet_name='CheckBook')

FCfinder_data.head()
CB_data.head()

# get the unique values of the column 'Station' from both dataframes
FCfinder_stations = FCfinder_data['Station'].unique()
CB_stations = CB_data['Station'].unique()

# get the common values of the column 'Station' from both dataframes as list
common_stations = list(set(FCfinder_stations) & set(CB_stations))
# sort the list
common_stations.sort()
print(common_stations)

# compare timeseries data for the common stations from both dataframes by plotting the data together
import matplotlib.pyplot as plt
import numpy as np
# loop through the common stations
for station in common_stations:
    # get the data for the station from both dataframes
    FCfinder_station_data = FCfinder_data[FCfinder_data['Station'] == station]
    CB_station_data = CB_data[CB_data['Station'] == station]
    # plot the data
    plt.figure()
    # change plot size
    plt.figure(figsize=(12, 6))
    plt.plot(FCfinder_station_data['Date'], FCfinder_station_data['Layer2'], label='FCfinder')

    plt.plot(CB_station_data['Date'], CB_station_data['Layer2'], label='CheckBook')
    plt.xlabel('Date')
    plt.ylabel('AVG')
    # y limit to 20 - 50
    plt.ylim(20, 60)
    plt.title(f'Station {station}')
    plt.legend()
    plt.show()
```
